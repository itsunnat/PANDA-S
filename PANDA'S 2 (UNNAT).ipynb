JOINS

```python
#Creating DF1 with dictionary

dict1={"name":["Pooja","Ketan","Jinal","Hiren","Tina","Leena"],"SQL":[91,92,93,94,95,96]}
DF1=pd.DataFrame(dict1)
DF1
```

```python
dict2={"name":["Mohan","Pooja","Tina","Payal","Reena","Jinal","Leena","Umesh"],"Python":[81,82,83,84,85,86,87,88]}
DF2=pd.DataFrame(dict2)
DF2
```

```python
#MERGING BOTH DF ON THE BASIS COMMON COLUMNS NAME

DF3=pd.merge(DF1,DF2,on="name")
DF3
#by default inner join will get perform on the basis of common column 
```

```python
#Full O+uter join
#match and unmatched data from both DF

DF4=pd.merge(DF1,DF2,on="name",how="outer")
DF4
#In front of unmatched data we can see Nan Values
```

```python
#LEFT JOIN

DF5=pd.merge(DF1,DF2,on="name",how="left")
DF5
#it consider DF1 as a left because we write it first on syntax
```

```python
#RIGHT JOIN

DF6=pd.merge(DF1,DF2,on="name",how="right")
DF6
```
